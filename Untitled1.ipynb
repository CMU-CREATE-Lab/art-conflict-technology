```python
from shapely.wkb import loads
from shapely.geometry import mapping, shape, Point
from random import uniform

def randomPoint(geom):
    poly = shape(geom)
    bbox = poly.bounds
    l,b,r,t = bbox
    while True:
        point = Point(uniform(l,r),uniform(t,b))
        if point is None:
            break
        if poly.contains(point):
            break
    return point.__geo_interface__['coordinates']

def make_point(point):
    geometry = {}
    geometry['type'] = "Point"
    geometry['coordinates'] = [point[0], point[1]]
    return geometry

def make_properties(race, race_color):
    properties = {}
    properties['race'] = race
    properties['race_color'] = race_color
    return properties

def make_feature(point, properties):
    feature = {}
    feature['type'] = 'Feature'
    feature['geometry'] = point
    feature['properties'] = properties
    return feature

race_colors = {
    'white': '25, 75, 255, 1',
    'black': '20, 138, 9, 1',
    'asian': '227, 30, 30, 1',
    'hispanic': '255, 170, 0, 1',
    'other': '137, 90, 68, 1'
}
```

```python
import random
block = json_data['features'][1000]
race_row = race_csv_data[block['properties']['GEOID10']]
def get_points_per_row(block, race_row):
    features = []
    total_all = int(race_row['H7Z001'])
    total_white = int(race_row['H7Z003'])
    total_black = int(race_row['H7Z004'])
    total_asian = int(race_row['H7Z006'])
    total_hispanic = int(race_row['H7Z010'])
    total_other = int(race_row['H7Z005']) + int(race_row['H7Z007']) + int(race_row['H7Z008']) + int(race_row['H7Z009'])
    if (total_all != total_white + total_black + total_asian + total_hispanic + total_other):
        print "ERROR: block id %s" % block['properties']['GEOID10']
    else:
        for i in range(0, total_white):
            geom = randomPoint(block['geometry'])
            point = make_point(geom)
            properties = make_properties('white', race_colors['white'])
            feature = make_feature(point, properties)
            features.append(feature)
        for i in range(0, total_black):
            geom = randomPoint(block['geometry'])
            point = make_point(geom)
            properties = make_properties('black', race_colors['black'])
            feature = make_feature(point, properties)
            features.append(feature)
        for i in range(0, total_asian):
            geom = randomPoint(block['geometry'])
            point = make_point(geom)
            properties = make_properties('asian', race_colors['asian'])
            feature = make_feature(point, properties)
            features.append(feature)
        for i in range(0, total_hispanic):
            geom = randomPoint(block['geometry'])
            point = make_point(geom)
            properties = make_properties('hispanic', race_colors['hispanic'])
            feature = make_feature(point, properties)
            features.append(feature)
        for i in range(0, total_other):
            geom = randomPoint(block['geometry'])
            point = make_point(geom)
            properties = make_properties('other', race_colors['other'])
            feature = make_feature(point, properties)
            features.append(feature)
    random.shuffle(features)
    return features
```

```python
# Generate a GeoJSON file of victims from XML
from xml.dom import minidom
import json, os

def xml2geojson(filename):
    name, ext = os.path.splitext(filename)
    geojsonfile = name + '.geojson'
    geojson = {}
    geojson['type'] = "FeatureCollection"
    geojson['features'] = []
    #reader = csv.DictReader(open(filename, 'rb'));
    xmldoc = minidom.parse("capture/genxml_victims.xml")
    markerlist = xmldoc.getElementsByTagName('marker') 
    for marker in markerlist:
        feature = {}
        feature['type'] = 'Feature'
        feature['geometry'] = {}
        feature['geometry']['type'] = 'Point'
        feature['geometry']['coordinates'] = [float(marker.attributes['lng'].value), float(marker.attributes['lat'].value)]
        feature['properties'] = {}
        keys = marker.attributes.keys()
        for key in keys:
            feature['properties'][key] = marker.attributes[key].value
        geojson['features'].append(feature)
    with open(geojsonfile, 'w') as outfile:
        json.dump(geojson, outfile, indent=4, separators=(',', ': '))
        
```
